```python
%env CUDA_LAUNCH_BLOCKING=1
import torch
from torch import nn
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
import pandas as pd
from chess_model.db.management.commands.train import xNNChess, ChessEvalDataset
```

```python
%%capture
from tqdm.notebook import tqdm
tqdm().pandas()
```

```python
model = xNNChess()
model.to('cuda')
```

```python
evals = ChessEvalDataset("eval100ms470k.csv")
evals
```

```python
dataloader = DataLoader(evals, batch_size=64,
                        shuffle=True, num_workers=0)
```

```python

loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=5e-3)
size = len(dataloader.dataset)
        
currs = []
losses = []
    
for batch, (X, y) in enumerate(dataloader):
    #print(X.shape)
    X = X.to('cuda')
    y = y.to('cuda')
    pred = model(X)
    #print(y.shape, pred.shape)
    loss = loss_fn(pred, y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    loss, current = loss.item(), batch * len(X)
    currs.append(current)
    losses.append(loss)
    
    if batch % 100 == 0:
        #print("Batch", batch)
        print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def gauss_filter(xs, ys, sigma):
    weights = norm.pdf(
        (xs.reshape(len(xs), 1)
        - xs.reshape(1, len(xs)))
        / sigma
    )
    return np.matmul(weights, ys) / weights.sum(axis=1)
```

```python
currs = np.array(currs)
smoothed_1000 = gauss_filter(currs, losses, 1000)
smoothed_5000 = gauss_filter(currs, losses, 5000)
```

```python
plt.plot(currs[:-1], smoothed_1000[:-1], 'lightskyblue')
plt.plot(currs[:-1], smoothed_5000[:-1], 'black')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.show()
```
